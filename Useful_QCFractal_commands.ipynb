# 2. load the collection beforehand 

```python
ds_mols = client.get_collection('OptimizationDataset', 'co2_beep')
ds_mols.status(collapse=False)
```

## 3. Loop through specifications and start computations:

```python
for spec in ds_mols.df.columns:
    c = ds_mols.compute(spec, tag='m_project')
    print(c)
```

## 4. Inspect Errors

```python
rr = ds_mols.get_record("CD2_01","mpwb1k-d3bj_def2-tzvp")
print(rr.get_error())
```

## 5. Restart All jobs (and change tags)

```python
new_tag = "m_project"

for spec in ds_mols.df.columns:
    for entry in ds_mols.df.index:
        rr = ds_mols.get_record(entry,spec)
        ct =  client.modify_tasks("modify", new_tag=new_tag, base_result=rr.id)
        cr =  client.modify_tasks("restart", base_result=rr.id)
        print(entry,spec,ct,cr)
```
