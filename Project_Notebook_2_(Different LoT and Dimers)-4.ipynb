# calculating the BSSE by hand (bovolenta paper)

```python
ds_mols=client.get_collection('OptimizationDataset', 'c2h2_w12_1')
ds_mols.status(collapse=False)
```

```python
def BSSE(dimer,molecule_1,molecule_2):
    
    return dimer - (molecule_1+molecule_2)

    
```

```python
results_df['BSSE_CO2_CO']=BSSE(results_df['CO2_CO'],results_df['CO2'],results_df['CO'])
results_df['BSSE_CO2_CO2']=BSSE(results_df['CO2_CO2'],results_df['CO2'],results_df['CO2'])
results_df['BSSE_H2O_H2O']=BSSE(results_df['H2O_H2O'],results_df['H2O'],results_df['H2O'])


```
