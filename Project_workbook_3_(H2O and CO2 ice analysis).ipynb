```python
def string_map(lists):
    n=len(lists)
    number_map=[None] *n
    for i in range(len(lists)):    
        if lists[i] == 'cd1_01':
            number_map[i]='monomer'
        elif lists[i] == 'cd2_01':
            number_map[i]='dimer'
        elif lists[i] == 'cd3_01':
            number_map[i]='trimer (barrel)'
        else:
            number_map[i]='trimer (cyclic)'
    return number_map


```

```python
#1.00000 kcal/mol = 503.22272K
H2O_binding = ['be_H2O_CD1_01_MPWB1K-D3BJ','be_H2O_CD2_01_MPWB1K-D3BJ','be_H2O_CD3_01_MPWB1K-D3BJ','be_H2O_CD3_02_MPWB1K-D3BJ']

list_of_BE_values_H2O=[] 
list_of_BE_values_H2O_nocp=[]
surfaces_CO2=[]
CO2_surface_colors = {'monomer': 'red', 'dimer': 'green',
                      'trimer (barrel)':'blue','trimer (cyclic)': 'orange'}


for i in H2O_binding:
    
    #Getting the BE values for each binding interaction 
    
    ds_be= client.get_collection('ReactionDataset', i)
    BE=ds_be.get_values(stoich='default',method = 'WPBE-D3BJ').values
    BE_nocp=ds_be.get_values(stoich='be_nocp',method = 'WPBE-D3BJ').values
    # Creating a list of corresponding CO2 surface on which the bindings took place (for graphic purposes)
    surface_CO2 = [(f'{i.split("_")[1]}_{i.split("_")[2]}')
                            for i in ds_be.get_values(stoich='be_nocp').index]

    # Appending the BE (CP and no CP correction) and the surface names
    #('monomer','dimer' etc) to their respective lists
    
    list_of_BE_values_H2O=np.append(list_of_BE_values_H2O,BE*-503.22272)
    list_of_BE_values_H2O_nocp=np.append(list_of_BE_values_H2O_nocp,BE_nocp*-503.22272)
    surfaces_CO2=np.append(surfaces_CO2,surface_CO2)
    surface_name_CO2 = string_map(surfaces_CO2)

# Creating a DF to store all of the BE values
H2O_df=pd.DataFrame()

H2O_df['CP corrected']=list_of_BE_values_H2O
H2O_df['Non CP corrected']=list_of_BE_values_H2O_nocp
H2O_df['surface_CO2']= surface_name_CO2
H2O_df['color'] = H2O_df['surface_CO2'].map(CO2_surface_colors)
# Creating a cutoff of 50K to get of of structures that hve not binded
H2O_df=H2O_df[H2O_df['CP corrected']>= 50]


# Seaborn boxplot of BE values
sns.boxplot(data=H2O_df, color="lightblue",whis=(0, 100),zorder=1)

# Overlaying a scatter plot on the boxplot labelled by the CO2 binding surface (CP and no CP)


for surface, color in zip(H2O_df['surface_CO2'].unique(), H2O_df['color'].unique()):
    # Scatter for 'CP corrected'
    plt.scatter(np.zeros(len(H2O_df[H2O_df['surface_CO2'] == surface])), 
                H2O_df[H2O_df['surface_CO2'] == surface]['CP corrected'], 
                color=color, label=surface if surface not in plt.gca().get_legend_handles_labels()[1] else "", zorder=2)

    # Scatter for 'Non CP corrected'
    plt.scatter(np.ones(len(H2O_df[H2O_df['surface_CO2'] == surface])), 
               H2O_df[H2O_df['surface_CO2'] == surface]['Non CP corrected'], 
               color=color, label=surface if surface not in plt.gca().get_legend_handles_labels()[1] else "", zorder=2)


"""

Loop Through Unique Surface Types: The code loops through each unique surface type and its corresponding color. This ensures that each surface type is represented only once in the legend.

Conditional Labeling: The label parameter in the scatter() function checks if the surface type is already in the legend using plt.gca().get_legend_handles_labels(). If it is, it assigns an empty string to avoid duplicate entries.

Creating the Legend: After plotting, plt.legend() is called to create the legend with the title "CO2 Surface Types".

Plot Customization: The x-ticks, y-label, and title are added for clarity.

"""
    
    
    
surface_counts = H2O_df['color'].value_counts()
plt.title("$E_{b}$ distribution for $H_{2}O$ on $CO_{2}$")
plt.ylabel("$E_{b} (K)$")
plt.xlabel("$H_{2}O$")
plt.ylim(0, max(H2O_df['Non CP corrected']*1.5))
plt.legend(title='CO2 surfaces',loc='upper center')
plt.show()






```

## BE values for CH3

```python
#1.00000 kcal/mol = 503.22272K
CH3_binding = ['be_CH3_CD1_01_MPWB1K-D3BJ','be_CH3_CD2_01_MPWB1K-D3BJ','be_CH3_CD3_01_MPWB1K-D3BJ','be_CH3_CD3_02_MPWB1K-D3BJ']

list_of_BE_values_CH3=[] 
list_of_BE_values_CH3_nocp=[]
surfaces_CO2=[]
CO2_surface_colors = {'monomer': 'red', 'dimer': 'green',
                      'trimer (barrel)':'blue','trimer (cyclic)': 'orange'}


for i in CH3_binding:
    
    #Getting the BE values for each binding interaction 
    
    ds_be= client.get_collection('ReactionDataset', i)
    BE=ds_be.get_values(stoich='default',method = 'WPBE-D3BJ').values
    BE_nocp=ds_be.get_values(stoich='be_nocp',method = 'WPBE-D3BJ').values
    # Creating a list of corresponding CO2 surface on which the bindings took place (for graphic purposes)
    surface_CO2 = [(f'{i.split("_")[1]}_{i.split("_")[2]}')
                            for i in ds_be.get_values(stoich='be_nocp').index]

    # Appending the BE (CP and no CP correction) and the surface names
    #('monomer','dimer' etc) to their respective lists
    
    list_of_BE_values_CH3=np.append(list_of_BE_values_CH3,BE*-503.22272)
    list_of_BE_values_CH3_nocp=np.append(list_of_BE_values_CH3_nocp,BE_nocp*-503.22272)
    surfaces_CO2=np.append(surfaces_CO2,surface_CO2)
    surface_name_CO2 = string_map(surfaces_CO2)

# Creating a DF to store all of the BE values
CH3_df=pd.DataFrame()

CH3_df['CP corrected']=list_of_BE_values_CH3
CH3_df['Non CP corrected']=list_of_BE_values_CH3_nocp
CH3_df['surface_CO2']= surface_name_CO2
CH3_df['color'] = CH3_df['surface_CO2'].map(CO2_surface_colors)
# Creating a cutoff of 50K to get of of structures that hve not binded
CH3_df=CH3_df[CH3_df['CP corrected']>= 50]


# Seaborn boxplot of BE values
sns.boxplot(data=CH3_df, color="lightblue",whis=(0, 100),zorder=1)

# Overlaying a scatter plot on the boxplot labelled by the CO2 binding surface (CP and no CP)


for surface, color in zip(CH3_df['surface_CO2'].unique(), CH3_df['color'].unique()):
    # Scatter for 'CP corrected'
    plt.scatter(np.zeros(len(CH3_df[CH3_df['surface_CO2'] == surface])), 
                CH3_df[CH3_df['surface_CO2'] == surface]['CP corrected'], 
                color=color, label=surface if surface not in plt.gca().get_legend_handles_labels()[1] else "", zorder=2)

    # Scatter for 'Non CP corrected'
    plt.scatter(np.ones(len(CH3_df[CH3_df['surface_CO2'] == surface])), 
               CH3_df[CH3_df['surface_CO2'] == surface]['Non CP corrected'], 
               color=color, label=surface if surface not in plt.gca().get_legend_handles_labels()[1] else "", zorder=2)

    
    
surface_counts = CH3_df['color'].value_counts()
plt.title("$E_{b}$ distribution for $CH_{3}$ on $CO_{2}$")
plt.ylabel("$E_{b} (K)$")
plt.xlabel("$CH_{3}$")
plt.ylim(-20, max(CH3_df['Non CP corrected']*1.5))
plt.legend(title='CO2 surfaces',loc='upper center')
plt.show()






```

```python
import numpy as np




def get_xyz_positions(molecule):
    
    ds_mols  = client.get_collection("OptimizationDataset", "project_molecules")
    r=ds_mols.get_record(molecule,'mpwb1k-d3bj_def2-tzvp')                    

    a=(r.get_final_molecule().pretty_print()).split('\n')
    data = [s.split() for s in a 
         if  s.strip() and not s.startswith(('Geometry', 'Center', '-'))][3:]
    cleaned_data=[[d[0]]+ list(map(float,d[1:]))for d in data]
        
    return cleaned_data
    

def calculate_moments_of_inertia(atoms, coordinates):
    """
    Calculate the principal moments of inertia for a molecule.

    Parameters:
    atoms : list of tuples
        A list of tuples where each tuple contains the atomic symbol and mass of an atom.
    coordinates : numpy array
        A 2D numpy array where each row represents the (x, y, z) coordinates of an atom.

    Returns:
    numpy array
        The principal moments of inertia.
    """
    # Calculate the center of mass
    total_mass = sum(mass for _, mass,charge in atoms)
    coordinates = np.array([coordinates[i][1:] for i in range(len(coordinates))])
    center_of_mass = sum(mass * coordinates[i] for i, (_, mass,charge) in enumerate(atoms)) / total_mass
    
    # Translate coordinates to the center of mass
    translated_coords = np.array([coordinates[i] - center_of_mass for i, (_, mass,charge) in enumerate(atoms)]) 
    dipole_moment = np.zeros(3)
    for (atom, mass, charge), position in zip(atoms, coordinates):
        coord_com = position - center_of_mass    
        dipole_moment+= (charge * coord_com)*1e-10
    
    #dipole_moment = sum(charge*translated_coords[i] for i ,(_,mass,charge) in enumerate(atoms)) *1e-10
    #print(np.array([coordinates[i] for i, (_, mass,charge) in enumerate(atoms)]))
    # Initialize the inertia tensor
    inertia_tensor = np.zeros((3, 3))

    # Calculate the inertia tensor
    for i, (_, mass,charge) in enumerate(atoms):
        x, y, z = translated_coords[i]
        inertia_tensor[0, 0] += mass * (y**2 + z**2)
        inertia_tensor[1, 1] += mass * (x**2 + z**2)
        inertia_tensor[2, 2] += mass * (x**2 + y**2)
        inertia_tensor[0, 1] -= mass * x * y
        inertia_tensor[0, 2] -= mass * x * z
        inertia_tensor[1, 2] -= mass * y * z

    # Symmetrize the inertia tensor
    inertia_tensor[1, 0] = inertia_tensor[0, 1]
    inertia_tensor[2, 0] = inertia_tensor[0, 2]
    inertia_tensor[2, 1] = inertia_tensor[1, 2]

    # Calculate the principal moments of inertia
    principal_moments = np.linalg.eigvalsh(inertia_tensor)

    return principal_moments,total_mass,dipole_moment







molecules = {
    "CO": [("C", 12.011, 0.40), ("O", 15.999, -0.40)],  # Added partial charges
    "CH3OH": [("C", 12.011, 0.20), ("H", 1.008, +0.33), ("H", 1.008, +0.33), ("H", 1.008, +0.33), ("O", 15.999, -0.60), ("H", 1.008, +0.33)],
    "H2CO": [("H", 1.008, +0.33), ("H", 1.008, +0.33), ("C", 12.011, +0.40), ("O", 15.999, -0.66)],
    "H2": [("H", 1.008, 0.00), ("H", 1.008, 0.00)],
    "NH3": [("N", 14.007, -1.00), ("H", 1.008, +0.33), ("H", 1.008, +0.33), ("H", 1.008, +0.33)],
    "OH": [("O", 15.999, -0.66), ("H", 1.008, +0.33)],
    "CH4": [("C", 12.011, 0.00), ("H", 1.008, +0.25), ("H", 1.008, +0.25), ("H", 1.008, +0.25), ("H", 1.008, +0.25)],
    "CH2OH": [("C", 12.011, +0.20), ("H", 1.008, +0.33), ("H", 1.008, +0.33), ("O", 15.999, -0.60), ("H", 1.008, +0.33)],
    "CH3O": [("C", 12.011, +0.20), ("H", 1.008, +0.33), ("H", 1.008, +0.33), ("H", 1.008, +0.33), ("O", 15.999, -0.60)],
    "CH3": [("C", 12.011, 0.00), ("H", 1.008, +0.25), ("H", 1.008, +0.25), ("H", 1.008, +0.25)],
    "H2S": [("H", 1.008, +0.34), ("H", 1.008, +0.34), ("S", 32.06, -0.68)],
    "HCl": [("H", 1.008, +0.50), ("Cl", 35.45, -0.50)],
    "HCOOH": [("H", 1.008, +0.33), ("C", 12.011, +0.40), ("O", 15.999, -0.66), ("O", 15.999, -0.66), ("H", 1.008, +0.33)],
    "HCO": [("H", 1.008, +0.33), ("C", 12.011, +0.40), ("O", 15.999, -0.66)],
    "HF": [("H", 1.008, +0.50), ("F", 18.998, -0.50)],
    "HNC": [("H", 1.008, +0.50), ("N", 14.007, -0.50), ("C", 12.011, +0.50)],
    "NH2": [("N", 14.007, -1.00), ("H", 1.008, +0.33), ("H", 1.008, +0.33)],
    #"NHCH2": [("N", 14.007, -1.00), ("H", 1.008, +0.33), ("C", 12.011, +0.40), ("H", 1.008, +0.33), ("H", 1.008, +0.33)],
    "C2H2": [("C", 12.011, 0.00), ("C", 12.011, 0.00), ("H", 1.008, +0.25), ("H", 1.008, +0.25)],
    "H2O": [("H", 1.008, +0.33), ("H", 1.008, +0.33), ("O", 15.999, -0.66)],
    "N2": [("N", 14.007, 0.00), ("N", 14.007, 0.00)],
    "HCN": [("H", 1.008, +0.50), ("C", 12.011, +0.40), ("N", 14.007, -0.90)],
    "CO2": [  # Carbon Dioxide
        ("C", 12.011, 0.00),  # Partial charge on Carbon
        ("O", 15.999, -0.50),  # Partial charge on Oxygen 1
        ("O", 15.999, -0.50)   # Partial charge on Oxygen 2
    ]
}






list_of_moments_of_inertia=[]
for molecule,atom in molecules.items():
    atoms = atom
    coordinates = get_xyz_positions(molecule)
    #print(f'{molecule}:{np.array([coordinates[i][1:] for i in range(len(coordinates))])}')
    moments_of_inertia = calculate_moments_of_inertia(atoms, coordinates)[0]
    mass =calculate_moments_of_inertia(atoms, coordinates)[1] 
    dipole_moment=((calculate_moments_of_inertia(atoms, coordinates)[2])*1.6e-19)/(3.336e-29)
    #print(f'{molecule}:{dipole_moment}')
    d= {'molecule':molecule,'Ix':moments_of_inertia[0],'Iy':moments_of_inertia[1],
        'Iz':moments_of_inertia[2],'mass (amu)': mass,
        '$$\mu_{x}$$':dipole_moment[0],'$$\mu_{y}$$':dipole_moment[1],
       '$$\mu_{z}$$':dipole_moment[2]}
    list_of_moments_of_inertia=np.append(list_of_moments_of_inertia,d)
    
list_of_moments_of_inertia
    #print(f"Principal moments of inertia (amu): for {molecule} ", moments_of_inertia)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Functions
# def temperature(t, T0, beta):
#     """ Calculate temperature at time t. """
#     return T0 + beta * t

def rate_constant(T, k0, E, R):
    """ Calculate the rate constant at temperature T. """
    return k0 * np.exp(-E / (R * T))

def integrand(T, theta, k0, E, R, beta):
    """ Calculate the rate of change of coverage at time t. """
#     T = temperature(t, T0, beta)
    k = rate_constant(T, k0, E, R)
    return -(k * theta)/beta

# Parameters for the TPD simulation
R = 8.314  # Universal gas constant in J/(mol*K)

# Dictionaries to store data for each temperature range
temp_ranges = {
    "0-50 K": [],
    "51-100 K": [],
    "101-150 K": [],
    "151-200 K": []
}

for index, row in TPD_trimer_df.iterrows():
    k0 = row['pre exponent factor (s^-1)']  # Pre-exponential factor in 1/s
    E_k = row['Binding Energy (K)']  # Activation energy in K
    E = 8.314 * E_k  # Activation energy in J/mol
    T0 = row['Initial temp (K)']  # Initial temperature in K
    beta = row['heating rate (Ks^-1)']  # Heating rate in K/s
    theta0 = row['Initial coverage']  # Initial coverage
        
    # Time array
#       if row['molecule']!='HF':
    time = np.linspace(0, 10000, 10000)  # Time from 0 to ... seconds
#         else:
#             time = np.linspace(0, 13000, 10000)
        
        # Temperature array
    T_end=200
        
    temperatures = np.linspace(T0,T_end,int((T_end-T0)/beta))
        
        # Coverage array
    coverage = np.zeros_like(temperatures)
    coverage[0] = theta0

        # Numerical integration using Euler's method
    dt = time[1] - time[0]  # Time step
    dT = temperatures[1] - temperatures[0]
        
    for i in range(1, len(temperatures)):
        coverage[i] = coverage[i-1] + integrand(temperatures[i-1], coverage[i-1], k0, E, R, beta) * dT
        if coverage[i] < 0:
            coverage[i] = 0

        # Calculate desorption rate
    desorption_rate = -np.gradient(coverage, dT)
    

        # Find peak temperature
    max_x = temperatures[desorption_rate.argmax()]
    max_y = max(desorption_rate)
    TPD_trimer_df.loc[index, 'Peak Temp (K)'] = max_x
    TPD_trimer_df.loc[index, 'peak desorption rate (1/s)']=max_y
        
    if row['Surface'] ==r'$$CO_{2}$$ (trimer)' or row['Surface'] ==r'$$CO_{2}$$':        
        # Assign to the correct temperature range
        if 0 <= max_x <= 50:
            temp_ranges["0-50 K"].append((temperatures, desorption_rate, row['molecule'], max_x))
        elif 51 <= max_x <= 100:
            temp_ranges["51-100 K"].append((temperatures, desorption_rate, row['molecule'], max_x))
#         elif 101 <= max_x <= 150:
#             temp_ranges["101-150 K"].append((temperatures, desorption_rate, row['molecule'], max_x))
#         elif 151 <= max_x <= 200:
#             temp_ranges["151-200 K"].append((temperatures, desorption_rate, row['molecule'], max_x))

# Plot results in 4 separate subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=False, sharey=False)
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

# Define subplot titles based on the temperature ranges
range_labels = ["0-50 K", "51-100 K"]

for i, label in enumerate(range_labels):
    ax = axes[i]
    
    if temp_ranges[label]:  # Check if there is data for this range
        for temperatures, desorption_rate, molecule, peak_temp in temp_ranges[label]:
            mask = desorption_rate>= 1e-5
            if molecule !='$H_{2}$':
                # Get the corresponding surface type from the DataFrame
                surface_type = TPD_trimer_df[
                    (TPD_trimer_df['molecule'] == molecule) & 
                    (TPD_trimer_df['Peak Temp (K)'].round(1) == round(peak_temp, 1))
                ]['Surface'].values[0]
                
                # Set line style based on surface type
                line_style = '--' if surface_type == r'$$CO_{2}$$' else '-'
                
                ax.plot(temperatures[mask], desorption_rate[mask], 
                       linestyle=line_style,
                       label=f'{molecule} (Peak: {peak_temp:.1f}K)')
    
    ax.set_xlabel('Temperature (K)',fontsize=20)
    ax.set_ylabel('Desorption Rate ($s^{-1}$)',fontsize=20)
    #ax.set_title(f'TPD Curves on CO2 trimer surface: {label}')
    ax.tick_params(axis='both', which='major', labelsize=18)
    plt.tick_params(axis='both', which='major', labelsize=18)
    plt.xlim(45,100)
    #plt.ylim(0,0.025)
    ax.legend()
    ax.grid(True)
```

## Get pre exponential factor using TST and redhead method

```python
import pandas as pd

h = 6.626e-34
k_b = 1.38e-23
A=1e-19
amu = 1.67e-27
I_conversion_factor=amu*1e-20

def translation_2D(T_peak,m):
    
    Lambda = h/np.sqrt(2*np.pi*m*amu*k_b*T_peak)
    q_trans = A/Lambda**2
    return q_trans

def rotation_2D(sym_factor, T_peak, I_y,I_z):
    term_1= np.sqrt(np.pi)/(sym_factor*(h**2))
    term_2 = 8*(np.pi**2)*k_b*T_peak
    term_3 = np.sqrt(I_y*I_z*I_conversion_factor**2)
    q_rot_2D = term_1*term_2*term_3
    return q_rot_2D

def rotation_3D(sym_factor, T_peak,I_x,I_y,I_z):
    term_1= np.sqrt(np.pi)/(sym_factor*(h**3))
    term_2 = (8*(np.pi**2)*k_b*T_peak)**(3/2)
    term_3 = np.sqrt(I_x*I_y*I_z*I_conversion_factor**3)
    q_rot_3D = term_1*term_2*term_3
    return q_rot_3D


def TST(df,sym_factor, T_peak,I_x,I_y,I_z,m):
    
    term_1 = (k_b*T_peak)/h
    term_2 = translation_2D(T_peak,m)
    if df =='yes':
        term_3 = rotation_2D(sym_factor,T_peak,I_y,I_z)
    else:
        term_3= rotation_3D(sym_factor, T_peak,I_x,I_y,I_z)
    nu_TST = term_1*term_2*term_3
    return nu_TST

# def err_TST(df,err_T_peak,nu_TST,T_peak):
#     if df == 'yes':
#         err_nu= (7/2)*(err_T_peak/T_peak)*nu_TST
#     else:
#         err_nu= (6/2)*(err_T_peak/T_peak)*nu_TST
#     return err_nu

def readhead(beta,T_peak, E_b):
    term_1 = beta/T_peak
    term_2 = np.exp((E_b/T_peak)+3.64)
    nu_redhead=term_1*term_2
    return nu_redhead




# Changing data types from objects to floats
TPD_trimer_df['Binding Energy (K)'].astype(float)
TPD_trimer_df['Peak Temp (K)'].astype(float)
# Joining the TPD_lit and I_values dfs
joined_df=pd.merge(df_I_values_resetted,TPD_trimer_df,on='molecule')
joined_df['Binding Energy (K)'].astype(float)
joined_df['Peak Temp (K)'].astype(float)
joined_df.dtypes
# Calculating nu for TST for each row
for index, row in joined_df.iterrows():
    #if row['Surface'] != r'$$CO_{2}$$':
    nu_TST=TST(row['2D-rotations?'],row[r'$\sigma$'],
                row['Peak Temp (K)'],row[r'$I_{x}$'],
                row[r'$I_{y}$'],row[r'$I_{z}$'],row['mass (amu)'])
        
    joined_df.loc[index, r'$\nu_{{TST}}$'] = nu_TST
        

# Calculating nu for Redhead method for each row       
    nu_redhead = readhead(row['heating rate (Ks^-1)'],
                            row['Peak Temp (K)'],
                            row['Binding Energy (K)'])
    
    joined_df.loc[index, r'$\nu_{{RH}}$'] = nu_redhead
        
    # Calculating the thermal length   
    joined_df.loc[index,r'$\Lambda$'] = h/np.sqrt(2*np.pi*
                                                    row['mass (amu)']*amu*
                                                    k_b*row['Peak Temp (K)'])
    #calculating the 2D/3D translational/rotational partition function
    joined_df.loc[index,r'$q_{{tr}}^{{2D}}$']   = translation_2D(row['Peak Temp (K)'],row['mass (amu)'])
    if row['2D-rotations?']=='yes':
        joined_df.loc[index,r'$q_{{rot}}^{{2D}}$'] = rotation_2D(row[r'$\sigma$'],row['Peak Temp (K)'],
                    row[r'$I_{y}$'],row[r'$I_{z}$'])
    else:
        joined_df.loc[index,r'$q_{{rot}}^{{3D}}$'] = rotation_3D(row[r'$\sigma$'],row['Peak Temp (K)'],
                    row[r'$I_{x}$'],row[r'$I_{y}$'],
                    row[r'$I_{z}$'])
        

joined_df
```

## Biscection Method to get the Sublimation Radius

```python
import numpy as np

h = 6.64e-34
k_b = 1.38e-23
A = 1e-19
AU = 1.5e11

def bisection_method(f, a, b, nu_TST, E_b, tol=1e-6, max_iter=10000, expansion_factor=1.05):
    """
    Find the root of f(x) using the Bisection Method.
    
    Parameters:
    f : function
        The function whose root is to be found.
    a, b : float
        The interval [a, b] where the root is located.
    nu_TST, E_b : float
        Additional parameters for the function f.
    tol : float, optional
        The tolerance for convergence (default is 1e-10).
    max_iter : int, optional
        The maximum number of iterations (default is 10000).
    expansion_factor : float, optional
        The factor by which to expand the interval if no sign change is detected (default is 1.5).
    
    Returns:
    root : float
        The approximate root of the function.
    """
    # Adjust the interval if no sign change is detected
    adjustment_count = 0
    while f(a, nu_TST, E_b) * f(b, nu_TST, E_b) >= 0 and adjustment_count < max_iter:
        a -= expansion_factor * (b - a)
        b += expansion_factor * (b - a)
        adjustment_count += 1
        print(f"Adjusting interval to [{a}, {b}]")

    if adjustment_count == max_iter:
        raise ValueError("Failed to find a valid interval with a sign change.")

    iteration = 0
    while (b - a) / 2 > tol and iteration < max_iter:
        c = (a + b) / 2  # Midpoint
        if f(c, nu_TST, E_b) == 0:
            error=(b-a)/2
            return c,error  # Found exact root
        elif f(a, nu_TST, E_b) * f(c, nu_TST, E_b) < 0:
            b = c  # Root is in the left half
        else:
            a = c  # Root is in the right half
        iteration += 1
        
    root = (a+b)/2
    error=(b-a)/2
       
    
    return root,error # Return the approximate root, error

# Example function (random polynomial equation)
def transcendental_radius_equation(R, nu_TST, E_b):
    T_0 = 200 
    R_0 = 1.5e11  # 1 AU
    mu = 2.34  # mean molecular mass
    m_p = 1.67e-27  # mass of proton 
    alpha = 1e-2  # alpha viscosity
    G = 6.67e-11  # Gravitational Constant
    M_star = 2e30

    phi1 = k_b * T_0 * np.sqrt(R_0)
    phi2 = ((mu * m_p) / (alpha * k_b * T_0)) * nu_TST * np.sqrt((G * M_star) / R_0)
    final = phi1 * np.log(phi2 * R) - E_b * np.sqrt(R)
    return final  

for index, row in joined_df.iterrows():
    
    # Interval where the root lies (must bracket a root)
    a, b = 1 * AU, 1e5 * AU
    nu_TST = row[r'$\nu_{{TST}}$']
    nu_redhead = row[r'$\nu_{{RH}}$']
    E_b = (row['Binding Energy (K)']*8.314)/6.02e23 # from K--> J/mol --> J
               
               

        # Finding the root using Bisection Method (root: TST, root_2: Redhead)
    try:
        root,error = bisection_method(transcendental_radius_equation, a, b, nu_TST, E_b)
        root_2,error_2 = bisection_method(transcendental_radius_equation, a, b, nu_redhead, E_b)
        joined_df.loc[index,r'$R_{s} (AU)$'] = root/AU
#       joined_df.loc[index,r'$\Delta R_{s} (AU)$'] = error/AU
        joined_df.loc[index,r'$R_{s,RH} (AU)$'] = root_2/AU
        print(f"Approximate root for {row['molecule']}: {root / AU:.6f}, {root_2 / AU:.6f}, {np.abs(root-root_2)*100/root} %")
            
    except ValueError as e:
        print(f"Error finding root for {row['molecule']}: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Functions
# def temperature(t, T0, beta):
#     """ Calculate temperature at time t. """
#     return T0 + beta * t

def rate_constant(T, k0, E, R):
    """ Calculate the rate constant at temperature T. """
    return k0 * np.exp(-E / (R * T))

def integrand(T, theta, k0, E, R, beta):
    """ Calculate the rate of change of coverage at time t. """
#     T = temperature(t, T0, beta)
    k = rate_constant(T, k0, E, R)
    return -(k * theta)/beta

# Parameters for the TPD simulation
R = 8.314  # Universal gas constant in J/(mol*K)

# Dictionaries to store data for each temperature range
temp_ranges = {
    "0-50 K": [],
    "51-100 K": [],
    "101-150 K": [],
    "151-200 K": []
}

color_map = {
    r'$$H_{2}O$$': 'blue',
    r'$$CO_{2}$$ (trimer)': 'red'
     
}

plt.figure(figsize=(12,10))
for index, row in joined_df.iterrows():
    if row['Surface'] !=r'$$CO_{2}$$' and row['molecule']==r'$H_{2}O$':
        k0 = row[r'$\nu_{{TST}}$']  # Pre-exponential factor in 1/s
        k1 = row[r'$\nu_{{RH}}$']
        E_k = row['Binding Energy (K)']  # Activation energy in K
        E = 8.314 * E_k  # Activation energy in J/mol
        T0 = row['Initial temp (K)']  # Initial temperature in K
        beta = row['heating rate (Ks^-1)']*0.001  # Heating rate in K/s
        theta0 = row['Initial coverage']  # Initial coverage
        beta2=10
        
        # Time array
#         if row['molecule']!='HF':
        time = np.linspace(0, 10000, 10000)  # Time from 0 to ... seconds
#         else:
#             time = np.linspace(0, 13000, 10000)
        
        # Temperature array
        T_end=200
        
        temperatures = np.arange(T0,T_end,(beta))
        
        # Coverage array
        coverage = np.zeros_like(temperatures)
        coverage_2 =np.zeros_like(temperatures)
        coverage[0] = theta0
        coverage_2[0]=theta0

        # Numerical integration using Euler's method
        dt = time[1] - time[0]  # Time step
        dT = temperatures[1] - temperatures[0]
        
        for i in range(1, len(temperatures)):
            coverage[i] = coverage[i-1] + integrand(temperatures[i-1], coverage[i-1], k0, E, R, beta) * dT
            if coverage[i] <= 0:
                coverage[i] = 0
                
        for i in range(1, len(temperatures)):
            coverage_2[i] = coverage_2[i-1] + integrand(temperatures[i-1], coverage_2[i-1], k1, E, R, beta) * dT
            if coverage_2[i] <= 0:
                coverage_2[i] = 0
        # Calculate desorption rate
        desorption_rate = -np.gradient(coverage, dT)
        desorption_rate_2 = -np.gradient(coverage_2, dT)

        # Find peak temperature
        max_x = temperatures[desorption_rate.argmax()]
        max_y = max(desorption_rate)
        max_x_2 = temperatures[desorption_rate_2.argmax()]
        max_y_2 = max(desorption_rate_2)

        surface_color = color_map.get(row['Surface'],'black')

    # Plot results for each initial coverage
        plt.plot(temperatures[desorption_rate>0], desorption_rate[desorption_rate>0],color=surface_color, label='TST')
        plt.plot(temperatures[desorption_rate_2>0], desorption_rate_2[desorption_rate_2>0],linestyle= '--',color=surface_color, label='Redhead')
        plt.annotate(f'{max_x:.1f} K',
                    xy=(max_x,max_y+0.001),
                     fontsize=18, color=surface_color)
        plt.annotate(f'{max_x_2:.1f} K',
                    xy=(max_x_2,max_y_2+0.001),
                     fontsize=18, color=surface_color)
        print(max_x)
        print(beta)
        print(temperatures)
        

plt.xlabel('Temperature (K)',fontsize=20)
plt.ylabel('Desorption Rate (1/s)',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=18)
plt.title(f'TPD Curves for $H_{2}O$',fontsize=20)
#plt.xlim(0,200)
#plt.ylim(0,0.025)
#plt.legend()
plt.grid(True)


plt.tight_layout()
plt.show()
```
