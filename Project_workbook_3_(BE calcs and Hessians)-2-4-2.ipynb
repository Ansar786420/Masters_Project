```python

ds_be = client.get_collection('ReactionDataset','be_H2O_CD1_01_MPWB1K-D3BJ')
dir(ds_be)

list(set([(f'{i.split("_")[1]}_{i.split("_")[2]}')for i in ds_be.get_values(stoich='be_nocp').index]))
    #(f'{i.split("_")[1]}_{i.split("_")[2]}')

```

```python
def string_map(lists):
    n=len(lists)
    number_map=[None] *n
    for i in range(len(lists)):    
        if lists[i] == 'cd1_01':
            number_map[i]='monomer'
        elif lists[i] == 'cd2_01':
            number_map[i]='dimer'
        elif lists[i] == 'cd3_01':
            number_map[i]='trimer (barrel)'
        else:
            number_map[i]='trimer (cyclic)'
    return number_map


```

```python
#1.00000 kcal/mol = 503.22272K
H2O_binding = ['be_H2O_CD1_01_MPWB1K-D3BJ','be_H2O_CD2_01_MPWB1K-D3BJ','be_H2O_CD3_01_MPWB1K-D3BJ','be_H2O_CD3_02_MPWB1K-D3BJ']

list_of_BE_values_H2O=[] 
list_of_BE_values_H2O_nocp=[]
surfaces_CO2=[]
CO2_surface_colors = {'monomer': 'red', 'dimer': 'green',
                      'trimer (barrel)':'blue','trimer (cyclic)': 'orange'}


for i in H2O_binding:
    
    #Getting the BE values for each binding interaction 
    
    ds_be= client.get_collection('ReactionDataset', i)
    BE=ds_be.get_values(stoich='default',method = 'MPWB1K-D3BJ').values
    BE_nocp=ds_be.get_values(stoich='be_nocp',method = 'MPWB1K-D3BJ').values
    # Creating a list of corresponding CO2 surface on which the bindings took place (for graphic purposes)
    surface_CO2 = [(f'{i.split("_")[1]}_{i.split("_")[2]}')
                            for i in ds_be.get_values(stoich='be_nocp').index]

    # Appending the BE (CP and no CP correction) and the surface names
    #('monomer','dimer' etc) to their respective lists
    
    list_of_BE_values_H2O=np.append(list_of_BE_values_H2O,BE*-503.22272)
    list_of_BE_values_H2O_nocp=np.append(list_of_BE_values_H2O_nocp,BE_nocp*-503.22272)
    surfaces_CO2=np.append(surfaces_CO2,surface_CO2)
    surface_name_CO2 = string_map(surfaces_CO2)

# Creating a DF to store all of the BE values
H2O_df=pd.DataFrame()

H2O_df['CP corrected']=list_of_BE_values_H2O
H2O_df['Non CP corrected']=list_of_BE_values_H2O_nocp
H2O_df['surface_CO2']= surface_name_CO2
H2O_df['color'] = H2O_df['surface_CO2'].map(CO2_surface_colors)
# Creating a cutoff of 50K to get of of structures that hve not binded
H2O_df=H2O_df[H2O_df['CP corrected']>= 50]


# Seaborn boxplot of BE values
sns.boxplot(data=H2O_df, color="lightblue",whis=(0, 100),zorder=1)

# Overlaying a scatter plot on the boxplot labelled by the CO2 binding surface (CP and no CP)


for surface, color in zip(H2O_df['surface_CO2'].unique(), H2O_df['color'].unique()):
    # Scatter for 'CP corrected'
    plt.scatter(np.zeros(len(H2O_df[H2O_df['surface_CO2'] == surface])), 
                H2O_df[H2O_df['surface_CO2'] == surface]['CP corrected'], 
                color=color, label=surface if surface not in plt.gca().get_legend_handles_labels()[1] else "", zorder=2)

    # Scatter for 'Non CP corrected'
    plt.scatter(np.ones(len(H2O_df[H2O_df['surface_CO2'] == surface])), 
               H2O_df[H2O_df['surface_CO2'] == surface]['Non CP corrected'], 
               color=color, label=surface if surface not in plt.gca().get_legend_handles_labels()[1] else "", zorder=2)


"""

Loop Through Unique Surface Types: The code loops through each unique surface type and its corresponding color. This ensures that each surface type is represented only once in the legend.

Conditional Labeling: The label parameter in the scatter() function checks if the surface type is already in the legend using plt.gca().get_legend_handles_labels(). If it is, it assigns an empty string to avoid duplicate entries.

Creating the Legend: After plotting, plt.legend() is called to create the legend with the title "CO2 Surface Types".

Plot Customization: The x-ticks, y-label, and title are added for clarity.

"""
    
    
    
surface_counts = H2O_df['color'].value_counts()
plt.title("$E_{b}$ distribution for $H_{2}O$ on $CO_{2}$")
plt.ylabel("$E_{b} (K)$")
plt.xlabel("$H_{2}O$")
plt.ylim(0, max(H2O_df['Non CP corrected']*1.5))
plt.legend(title='CO2 surfaces',loc='upper center')
plt.show()






```
