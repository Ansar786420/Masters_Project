## Get BE values for H2O on CD

```python

ds_be = client.get_collection('ReactionDataset','be_H2O_CD1_01_MPWB1K-D3BJ')
dir(ds_be)
ds_be.get_values(stoich='be_nocp')
```

```python
#1.00000 kcal/mol = 503.22272K
H2O_binding = ['be_H2O_CD1_01_MPWB1K-D3BJ','be_H2O_CD2_01_MPWB1K-D3BJ','be_H2O_CD3_01_MPWB1K-D3BJ','be_H2O_CD3_02_MPWB1K-D3BJ']

list_of_BE_values_H2O=[]
list_of_BE_values_H2O_nocp=[]
for i in H2O_binding:
    ds_be= client.get_collection('ReactionDataset', i)
    BE=ds_be.get_values(stoich='default',method = 'MPWB1K-D3BJ').values
    BE_nocp=ds_be.get_values(stoich='be_nocp',method = 'MPWB1K-D3BJ').values
    list_of_BE_values_H2O=np.append(list_of_BE_values_H2O,BE*503.22272)
    list_of_BE_values_H2O_nocp=np.append(list_of_BE_values_H2O_nocp,BE_nocp*503.22272)



H2O_df=pd.DataFrame()

H2O_df['CP corrected']=-list_of_BE_values_H2O
H2O_df['Non CP corrected']=-list_of_BE_values_H2O_nocp
H2O_df
#labels = [f"Point {i}" for i in range(len(-list_of_BE_values_H2O))]  # Example labels

# Seaborn boxplot
sns.boxplot(data=H2O_df, color="lightblue",whis=(0, 100))
for column_values in H2O_df['CP corrected']:
    plt.scatter(0, column_values, color='red',zorder=2)  # 0 is the x-position for Seaborn boxplot
    #plt.text(0.1, point, s='', fontsize=9, verticalalignment='center')
for column_values in H2O_df['Non CP corrected']:
    plt.scatter(1, column_values, color='blue',zorder=2)  # 0 is the x-position for Seaborn boxplot
    #plt.text(0.1, point, s='', fontsize=9, verticalalignment='center')

plt.title("$E_{b}$ distribution for $H_{2}O$ on $CO_{2}$")
plt.ylabel("$E_{b}$")
plt.xlabel("$H_{2}O$")

plt.show()






```

## Create a multi-boxplot

```python
## Create a Multi- boxplot 


BE_vals= np.array([-list_of_BE_values_H2O,-list_of_BE_values_CH4,-list_of_BE_values_CO,
                    -list_of_BE_values_C2H2,-list_of_BE_values_H2CO]) 

BE_df = pd.concat([H2O_df,CH4_df,CO_df,C2H2_df,H2CO_df],axis=1)
#BE_df
# Calculate row means
#BE_df["Row Mean"] = BE_df["Values"].apply(lambda x: sum(x) / len(x))


labels = [
    [f"Pt {i}" for i in range(len(BE_vals))] for group in BE_vals
]  # Create labels for each point in each group

# Create the boxplot
fig, ax = plt.subplots(figsize=(10, 6))
ax.boxplot(BE_vals, patch_artist=True, boxprops=dict(facecolor="lightblue"), widths=0.5)


                     
# Label points
for i, group in enumerate(BE_vals, start=1):
    for j, point in enumerate(group):
        plt.scatter(i, point,color='red',zorder=2)  # 0 is the x-position for Seaborn boxplot
        #ax.text(i + 0.05, point, s='', fontsize=8, verticalalignment="center")
        


# Customizations
ax.set_xticks(range(1, len(BE_vals) + 1))
ax.set_xticklabels(['$H_{2}O$','$CH_{4}$', '$CO$', '$C_{2}H_{2}$','$H_{2}CO$'])  # Labels for each boxplot
plt.title("Multi-Boxplot with Labelled Points")
plt.ylabel("$E_{b}$")

plt.show()
```
